Question 1:What is Dimensionality Reduction? Why is it important in machine learning?

Answer:
Dimensionality reduction is the process of reducing the number of input variables (features) in a dataset while retaining as much relevant information as possible.
It transforms data from a high-dimensional space to a lower-dimensional one.

Importance:

  - Removes noise and redundant features

  - Speeds up computation and training time

  - Improves model generalization and reduces overfitting

  - Enables easier visualization and interpretation of data

  - Helps algorithms perform better when “curse of dimensionality” exists

Question 2:Name and briefly describe three common dimensionality reduction techniques.

Answer:

- Principal Component Analysis (PCA): Converts correlated variables into a smaller number of uncorrelated principal components that capture maximum variance.

- t-Distributed Stochastic Neighbor Embedding (t-SNE):
Non-linear technique used mainly for visualization; preserves local structure of data in 2-D or 3-D.

- Linear Discriminant Analysis (LDA):
Supervised method that maximizes class separability by projecting data onto directions that best separate different classes.

Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.

Answer:

Clustering is an unsupervised learning technique that groups similar data points into clusters such that objects in the same cluster are more similar to each other than to those in other clusters.

- Popular clustering algorithms:

  - K-Means Clustering – partitions data into K clusters by minimizing within-cluster variance.

  - Hierarchical Clustering – builds a tree (dendrogram) of clusters without pre-specifying K.

  - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) – groups dense regions and detects outliers automatically.

Question 4: Explain the concept of anomaly detection and its significance.

Answer:

Anomaly detection identifies data points, events, or observations that deviate significantly from the normal behavior of the dataset.

Significance:
  - Detects fraud in transactions

  - Identifies machine or sensor faults early

  - Flags unusual user or network activity for cybersecurity

  - Maintains quality control in manufacturing

  - Useful in medical diagnosis for rare diseases.

Question 5: List and briefly describe three types of anomaly detection techniques.

Answer:

- Statistical Methods:
Assume data follows a probability distribution; anomalies are data points with low likelihood (e.g., z-score, Gaussian models).

- Machine Learning-Based Methods:
Use models like Isolation Forest, One-Class SVM, or Autoencoders that learn the normal pattern and detect deviations.

- Distance-Based / Density-Based Methods:
Identify anomalies based on their distance or local density compared to neighbors (e.g., Local Outlier Factor, KNN-based methods).

Question 6: What is time series analysis? Mention two key components of time series data.

Answer:

Time series analysis studies data points collected or recorded over time to identify meaningful statistics, trends, and patterns.

Key components:

  - Trend: Long-term increase or decrease in data values over time.

  - Seasonality: Repeating patterns that occur at fixed periods (e.g., daily, monthly).

Question 7: Describe the difference between seasonality and cyclic behavior in time series.

Answer:

| Feature            | Seasonality                            | Cyclic Behavior                   |
| ------------------ | -------------------------------------- | --------------------------------- |
| **Definition**     | Regular and predictable fluctuations   | Irregular, long-term fluctuations |
| **Time length**    | Fixed period (e.g., 12 months, 7 days) | Varies in duration                |
| **Example**        | Sales increase every December          | Business cycles in economy        |
| **Predictability** | Highly predictable                     | Less predictable                  |



Question 8:
Write Python code to perform K-Means clustering on a sample dataset.

Answer :

```python
#  K-Means clustering
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Generate sample data (3 clusters)
X, y_true = make_blobs(n_samples=500, centers=3, random_state=42)

# 2. Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# 3. Plot clusters
plt.figure(figsize=(10,8))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans,edgecolors='black', cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', label='Centroids')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# 4. Print basic results
print("Cluster Centers:\n", centers)
print("Inertia (Within-cluster sum of squares):", kmeans.inertia_)
```

Question 9:

What is inheritance in OOP? Provide a simple example in Python.

Answer:

Inheritance is an OOP concept where one class (child/subclass) inherits attributes and methods from another class (parent/superclass).
It promotes code reuse and hierarchical relationships.

```python
# Inheritance example
# parent class
class Animal:
    def speak(self):
        print("Animals make sounds.")

# Child class inherits from Animal
class Dog(Animal):
    def speak(self):
        print("Dog barks!")

# Create objects
a = Animal()
d = Dog()

a.speak()
d.speak()
```

Question 10:

How can time series analysis be used for anomaly detection?

Answer:

Time series analysis helps detect anomalies by modeling expected temporal patterns (trend, seasonality) and identifying deviations.

Approaches:

  - Statistical Models:
Fit ARIMA, SARIMA, or exponential smoothing to forecast expected values; anomalies are points with large residual errors.

  - Machine Learning / Deep Learning:
Use LSTM, Prophet, or Autoencoders to learn temporal dependencies and detect sudden spikes or drops.

  - Moving Window / Rolling Metrics:
Compute rolling mean and standard deviation; flag points outside defined thresholds.

Applications:

  - Network intrusion detection

  - Equipment failure prediction (IoT sensors)

  - Fraud detection in transaction streams

  - Detecting abnormal traffic or power consumption

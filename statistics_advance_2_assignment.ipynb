Question 1: What is hypothesis testing in statistics?

Answer: Hypothesis testing in statistics is a method used to make decisions or draw conclusions about a population based on sample data.

Example: Testing if a new medicine lowers blood pressure more than the current one.
 - H₀(Null hypothesis): New medicine has the same effect.
 - H₁(Alternate hypothesis): New medicine is more effective.


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:

Null Hypothesis (H₀):
  -  A statement that assumes there is no effect, no difference, or no relationship in the population.

  - It acts as the default or starting assumption in hypothesis testing.

  -  Example: “The average height of students in School A is equal to 160 cm.”

Difference:
- In Null hypothesis nothing unusual is happening and in Alternate hypothesis something unusual is happening.


Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

Answer: The significance level, often denoted by α (alpha), is a critical threshold in hypothesis testing that helps determine whether to reject the null hypothesis (H₀).

**Role in Hypothesis Testing:**

- Set before testing – Example: α = 0.05 means you are willing to accept a 5% chance of wrongly rejecting H₀.

- Compare with the p-value –

  -  If p ≤ α → reject H₀ (evidence is strong against H₀).

  -  If p > α → fail to reject H₀ (evidence is not strong enough).

- Determines critical value – In statistical tables, α helps decide the cut-off region for rejecting H₀.


Question 4: What are Type I and Type II errors? Give examples of each.

Answer:

Type I error : Rejecting the null hypothesis when it is actually true.
- Example:
 - In a COVID test: Declaring a healthy person as “infected.”
 - In a court trial: Convicting an innocent person.


Type II error : Failling to reject the null hypothesis when it is actually false.
- Example:
 - In a COVID test: Declaring an infected person as “healthy.”

 - In a court trial: Letting a guilty person go free.


Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.

Answer:

The **Z-test** is used when you have a **large sample size** (more than 30) and you **know the population standard deviation (σ)**. It relies on the **standard normal distribution** (Z-distribution), which has a fixed bell shape. It’s most suitable when your data is approximately normally distributed and you have enough observations to make the results stable.

The **T-test**, on the other hand, is used when you have a **small sample size** (30 or fewer) and you **do not know the population standard deviation**. It uses the **Student’s t-distribution**, which changes shape depending on the sample size — it’s wider for smaller samples to account for more uncertainty.

In short:

* **Z-test** → big sample, σ known.
* **T-test** → small sample, σ unknown.

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)



Answer:

```python
import numpy as np
import scipy.stats as stats
# given data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1,
               49.9, 50.8, 50.4, 48.9,50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3,
               50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
mu = 50 # hypothesized population mean claim by company
alpha = 0.05
# find sample statistics
s_mean= np.mean(sample_data)
s_std =np.std(sample_data)
# calculate z-statistics
n = len(sample_data)
z_stat = ( s_mean - mu)/(s_std/np.sqrt(n))
# calculate z-critical values
z_crit = stats.norm.ppf(1-alpha/2)# two tailed test
#output
print(f"The sample mean is:{s_mean:.2f}.")
print(f"The sample standerd deviation is: {s_std:.2f}.")
print(f"The value of the z-statistics is: {z_stat:.2f}. \nThe value of the z-critical is:{z_crit:.2f}. ")
#decision
if z_stat > z_crit:
  print("Reject the null hypothesis: the true mean is significantly different form the claim.")
else:
  print("Fail to reject the null hypothesis: no enough evidence against the claim.")


```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

Answer:  

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate normal distribution data
np.random.seed(10) # to reproducibility
mu = 50 # poplation mean
sigma = 5 # population standard deviation
n = 100 # sample size
data = np.random.normal(mu,sigma,n)

# Step 2: Calculate sample statistics
s_mean = np.mean(data) # sample mean
s_std = np.std(data,ddof = 1) # sample standard deviation

# Step 3: Calculate 95% confidence interval
alpha = 0.05
t_critical = stats.t.ppf(1-alpha/2, df = n-1)
mergin_error = ( t_critical* s_std)/ n**0.5
ci_lower = s_mean - mergin_error
ci_upper = s_mean + mergin_error


# Step 4: Output results
print(f"Sample Mean: {s_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 5: Plot histogram of the data
plt.hist(data, edgecolor = 'black', bins = 20, alpha= 0.8)
plt.title('Normal Distribution with 95 % CI')
plt.xlabel(' Value')
plt.ylabel('Frequency')
plt.axvline(s_mean,color = 'red', linestyle = '--',label = f"Mean = {s_mean:.2f}" )
plt.axvline(ci_lower,color = 'green', linestyle = '--',label = f"Lower CI = {ci_lower:.2f}" )
plt.axvline(ci_upper,color = 'green', linestyle = '--',label = f"Upper CI = {ci_upper:.2f}" )
plt.legend()
plt.show()


```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

Answer:  

```python
import numpy as np
import matplotlib.pyplot as plt

def zscore_histogram(data):
    data = np.array(data)

    # Mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Z-score calculation
    z_scores = (data - mean) / std_dev

    # Plot histogram
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title('Histogram of Standardized Data (Z-scores)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    return z_scores

# Example usage:
data = [12, 15, 14, 10, 18, 20, 15, 17, 14, 16,24,29,30,26]
z_values = zscore_histogram(data)
print("Z-scores:", z_values)
```

-  Interpret the Output:

 - Histogram center (Z = 0): Mean of data.

 - Right side of histogram (Z > 0): Above the mean.

 - Left side of histogram (Z < 0): Below the mean.

 - If the histogram looks like a bell curve, then data might be normally distributed.
